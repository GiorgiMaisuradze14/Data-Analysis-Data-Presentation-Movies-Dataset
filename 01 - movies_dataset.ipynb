## The best and the worst movies 1


```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
# მთავარი ცხრილის copy შევქმენით , იმ ველების მიხედვით რომლებსაც შემდგომში გამოვიყენებთ.
df_best = movies_df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best.head(3)
```

```python
# ცხრილს დავამატოთ 2 სვეტი. მოგება და რა დაბრუნდა.
df_best["profit_musd"] = movies_df.revenue_musd.sub(movies_df.budget_musd)
df_best["return"] = movies_df.revenue_musd.div(movies_df.budget_musd)
df_best.head(3)
```

```python
# სვეტებს შევუცვალოთ სახელები.
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
# Title იყოს პირველი სვეტი ინდექსის როლში.
df_best.set_index("Title", inplace = True)
df_best.head(3)
```

```python
# HTML ია ჩვეულებრივი , ლინკით კი რაღაც სურათზე გადავდივართ
df_best.iloc[0,0]
```

```python
# პატარა დემონსრაცია რა შეგვიძლია რომ გავაკეთოთ
subset = df_best.iloc[:5, :2]
subset
```

```python
# კოდი გადავიყვანოთ საიტის ფორმატში და სურათები გამოვაჩინოთ შედეგში.
HTML(subset.to_html(escape=False))
```

```python
# არ არის სარწმუნო , რადგან ერთი ადამინი აძლევს ხმას და კი ექნება 10.0 სხვა რამის მოფიქრება მოგვიწევს შემდეგში
df_best.sort_values(by = "Average Rating", ascending = False)
```

```python
# იგივე პრობლემა გვაქ აქ
df_best.sort_values(by = "ROI", ascending = False)
```

```python
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending = False)
```

```python
# nan ები შევცვალოთ 0 მნიშვნელობით
df_best['Budget'] = df_best['Budget'].fillna(0)
df_best['Votes'] = df_best['Votes'].fillna(0)
```

```python
df_best.info()
```

```python
# function --> best_worst(n, by, ascending = False, min_bud, min_votes):

def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
       
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), 
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

## Find your next Movie

```python
#Search 1: Science Fiction Action Movie with Bruce Willis (high Rating)
mask_genres = movies_df['genres'].str.contains('Action') & movies_df['genres'].str.contains('Science Fiction')
mask_actor = movies_df['cast'].str.contains('Bruce Willis')

#result
bruce = movies_df.loc[mask_actor & mask_genres, ["title", "poster_path", "vote_average"]].sort_values(by = "vote_average", ascending = False)
HTML(bruce.to_html(escape=False))   
```

```python
#Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (low runtime)
search_actor = movies_df['cast'].str.contains('Uma Thurman')
search_dirc = movies_df['director'].str.contains('Quentin Tarantino')

filter_films = movies_df.loc[search_actor & search_dirc , ('title','poster_path','runtime')].sort_values(by = "runtime").set_index("title")
HTML(filter_films.to_html(escape=False))  
```

```python
#Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (high Revenue)
pixar = movies_df['production_companies'].str.contains('Pixar')
mask_time = movies_df['release_date'].between("2010-01-01", "2015-12-31")

filter_films_1 = movies_df.loc[pixar & mask_time ,
                               ['title','poster_path','revenue_musd']].sort_values(by ='revenue_musd',ascending=False).set_index("title")
HTML(filter_films_1.to_html(escape=False))  
```

```python
#Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent)
mask_genres_1 = movies_df['genres'].str.contains('Action') | movies_df['genres'].str.contains('Thriller')
origin_lang = movies_df['original_language'].str.contains('en')
min_rait = movies_df['vote_average'] >= 7.5
mask_vote_co = movies_df['vote_count'] >= 10

#result
bruce = movies_df.loc[mask_genres_1 & origin_lang & min_rait & mask_vote_co , 
                      ["title", "poster_path", "genres", "vote_average", "vote_count", "release_date"]].sort_values(by = "release_date", ascending = False).set_index("title").head(20)
HTML(bruce.to_html(escape=False))   
```
